```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

Q1

```python
X = np.random.normal(1.5,0.5,(1000,))
mean = np.mean(X)
variance = np.var(X)
mad = stats.median_abs_deviation(X)
kurt = stats.kurtosis(X)
skew = stats.skew(X)
print("Mean  = {:f} \nVariance  = {:f} \nMAD  = {:f}  \nKurtosis  = {:f}  \nSkewness  = {:f}".format(mean,variance,mad,kurt,skew))
```

Q2

```python
xi = np.linspace(-4,4,1000)
X1 = stats.cauchy.pdf(xi,scale=1.5)
X2 = stats.norm.pdf(xi,scale=1.5)
plt.plot(xi,X1 , color = 'b')
plt.plot(xi,X2 , color = 'r')
plt.legend(["Cauchy" , "Normal"])
plt.show()
```

Q3

```python
xi2=np.linspace(-5,16,22)
xi3=np.linspace(-5,16,100)
X3 = stats.poisson.pmf(xi2,mu=5,loc=5)
X4 = stats.norm.pdf(xi3,loc=5)
plt.plot(xi2,X3 , color = 'b')
plt.plot(xi3,X4 , color = 'r')
plt.legend(["Poisson" , "Normal"])
plt.show()
```

```python
mean = np.array([.8920,.881,.8913,.9837,.8958])
error = np.array([.00044 , .009 , .00032,.00048, .00045])
w = error**-2
wMean = np.sum(mean*w)/np.sum(w)
deltaX = np.max(mean+error)/2 - np.min(mean+error)/2
meanError = deltaX/np.sqrt(len(w))
print("Weighted Mean Lifetime : {:f} \nWeighted Mean Uncertainity : {:f}".format(wMean ,meanError))
```
